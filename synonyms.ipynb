```python
from rake_nltk import Rake
import json
import re
from keyword_ import Keyword
from linggle import Linggle
```

```python
ex = json.load(open('ex.json','r'))
r = Rake(max_length=1)
```

```python
and_grams = {}

linggle = Linggle()

def ngramcount(query):
    return linggle[query]

for email_type, items in ex.items():
    text = []
    for categ, sentences in items.items():
        if categ != 'phrase':
            for sentence in sentences:
                text.append(sentence)
    #print(text[:100])
    #r.extract_keywords_from_sentences(text)
    print(email_type)
    #print(r.get_ranked_phrases()[:10])
    top_keywords = Keyword(text,[1])[:10]
    and_gram = []
    for idx, item in enumerate(ngramcount('%s and %s'%('/'.join(top_keywords), '/'.join(top_keywords)))):
        if idx < 3 or item[1] >= 10000:
            word1, word2 = item[0].split(' and ')
            and_gram.append((word1, word2, item[1]))
        else:
            break
    and_grams[email_type] = and_gram
```
